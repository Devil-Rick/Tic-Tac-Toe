```python
from tkinter import *
from tkinter import messagebox
```

#### Creating the basic background Window

```python
root = Tk()  # creating the gui window
root.configure(background = '#416fc4') # setting the background colour (configuring)
root.geometry("1350x720") # setting size of the window
root.title('Tic-Tac-Toe Game') # setting the title of calculator

#Changes to try
Game = LabelFrame(root , padx=15, pady=15 , bd = 10)
Score = LabelFrame(root , padx=15, pady=15 , bd = 10)
```

#### Creating the Required Functions
* Btn function

```python
# button on click func
    
def b_click(b):
    global clicked , count
    if b['text'] == ' ' and clicked == True:
            b['text'] = 'X'
            clicked = False
            count +=1
            who_won()
    elif b['text'] == ' ' and clicked == False:
            b['text'] = 'O'
            clicked = True
            count +=1
            who_won()
    else:
        messagebox.showerror('Tic-Tac-Toe' , 'Box already taken \n Pick another Box ....')
```

### Code for the Score window

```python
# req functions for scoring 

X_score = 0
O_score = 0

def score_X():
    global X_score 
    X_score +=1
    screen_X.set(X_score)

def score_O():
    global O_score
    O_score +=1
    screen_O.set(O_score)

def reset_score():
    global X_score , O_score
    X_score = 0
    screen_X.set(X_score)
    O_score = 0
    screen_O.set(X_score)
    reset()
    
```

```python
# contents for the score window

large_font = ('ds-digital' , 30)


screen_X = StringVar()
screen_O = StringVar()

text_X = Label(Score , text = 'Score X :' , font = large_font)
text_X.grid(row = 0 , column = 0)
display_X = Entry(Score , textvariable = screen_X , font = large_font , width = 15 )
display_X.grid(row = 0 , column = 1 , columnspan = 5 , ipady = 10 , padx = 40 , pady = 10)

text_O = Label(Score , text = 'Score O :' , font = large_font)
text_O.grid(row = 1 , column = 0)
display_O = Entry(Score , textvariable = screen_O , font = large_font , width = 15)
display_O.grid(row = 1 , column = 1, columnspan = 5 , ipady = 10 , padx = 40 , pady = 10)
```

### Main Menu and Starting the game 

```python
# creating the Main MENU 
my_menu = Menu(root)
root.config(menu = my_menu)

# creating options menu
my_options = Menu(my_menu , tearoff = False)
my_menu.add_cascade(label = 'Options' , menu = my_options)
my_options.add_command(label = 'Reset Score' , command = reset_score)
my_options.add_command(label = 'Reset Game' , command = reset)

Game.pack()
Game.place(x = 50 , y = 50)
Score.pack()
Score.place(x = 650 , y = 200)
# reseting and starting the game
reset_score()
reset()
root.mainloop()
```
